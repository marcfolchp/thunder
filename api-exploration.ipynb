## GameStats

```python
from euroleague_api.game_stats import GameStats

gamestats = GameStats(competition_code)
```

### GameStats.get_game_data()

```python
gamestats.get_game_data(season, game_code, endpoint="stats").columns
# endpoint puede ser report, stats o teamsComparison
# stats tiene datos de estadÃ­sticas del partido, y tiene una columna "local.players" y "road.players" con un diccionario con info de cada jugadores y sus estadÃ­sticas del partido.
```

```python
# cada endpoint no tiene columnas (osea a priori data) en comÃºn, cada endpoint da datos distintos del partido, no es que sea una manera de presentarlo.
cr = set(gamestats.get_game_data(season, game_code, endpoint="report").columns)
cs = set(gamestats.get_game_data(season, game_code, endpoint="stats").columns)
ctc = set(gamestats.get_game_data(season, game_code, endpoint="teamsComparison").columns)
comunes = cr & cs & ctc
comunes
```

### GameStats.get_game_report()

```python
# hace lo mismo que gamestats.get_game_data(season, game_code, endpoint="report") y lo mismo ocurre con el resto de mÃ©todos, son modificaciones de output del original, pero los datos son los mismos
gamestats.get_game_report(season, game_code)
```
